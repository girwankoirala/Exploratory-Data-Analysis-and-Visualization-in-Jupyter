```python
#Rename the experience level columns 
#SE – Senior Level/Expert
#MI – Medium Level/Intermediate
#EN – Entry Level
#EX – Executive Level

df = pd.read_csv('DataScienceSalaries.csv')


renaming = {
    'SE': 'Senior Level/Expert',
    'MI': 'Medium Level/Intermediate',
    'EN': 'Entry Level',
    'EX': 'Executive Level'
}

# this replaces the values across the entire DataFrame
df.replace(renaming, inplace=True)

#for displaying the updated DataFrame to confirm changes
print(df.head())


```

```python
# a python program to find out top 15 jobs. Make a bar graph of salary as well.         

Thetop15jobs = df['job_title'].value_counts().head(15)
print("Top 15 Jobs:\n", Thetop15jobs)
Thetop15jobs.plot(kind='bar', color='purple')  
plt.title('Top 15 Jobs')
plt.xlabel('Job Title')
plt.ylabel('Count') 
plt.xticks(rotation=65, ha='right')
plt.show()
```

```python
#job that has the highest salaries with Illustrate with bar graph.
highsalary=df['salary'].max()
total=df[df['salary']==highsalary]

#selecting top 15 highest paying job
top=df.sort_values(by='salary', ascending=False).head(15)
top


```

```python
#job that has the highest salaries with Illustrate with bar graph.	

top.plot(kind='bar',x='job_title',y='salary',color='orange')
plt.title('Top 15 Highest Paying Jobs')
plt.xlabel('Job Title')
plt.ylabel('Salary')
plt.xticks(rotation=65, ha='right', fontsize=9) 
plt.tight_layout()  
plt.show()
```

```python


# Group data by experience level and calculate average salary for each group

experience_salary = df.groupby('experience_level')['salary'].mean().sort_index()

# Plot a bar graph of salaries based on experience level
experience_salary.plot(kind='bar', color='grey')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary')
plt.xticks(rotation=65, ha='right')

plt.show()
```

```python
#	Write a Python program to show histogram and box plot of any chosen different variables. Use proper labels in the graph.

df['salary_in_usd'].hist(color='purple', edgecolor='black')
plt.title('The Histogram of Salary in usd')
plt.xlabel('salary in usd')
plt.ylabel('range')


```

```python
# Creating a box plot

#setting figure size
plt.figure(figsize=(8,8))
df.boxplot(column=['salary_in_usd']) 

# Adding gridlines for better readability
plt.grid(True)  
plt.title('The Box plot of Salary in usd')
plt.ylabel('range')
plt.show()  
```
